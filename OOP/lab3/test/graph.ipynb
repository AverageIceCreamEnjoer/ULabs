## Графики симметричного гиперболического распределения с параметрами $\nu, \mu\space и \space \lambda$

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../test321.csv", sep=",")
print(df.info())
plt.figure(figsize=(16, 9))
plt.plot(df['x'], df['f(x)'], linewidth=2,
         label='Теоретическое распределение смеси\nf(x) nu1=1, lambda1=2, mu1=10\nnu2=1, lambda2=2, mu2=10, p=0.5', alpha=0.8)
plt.scatter(df['xi'], df['f(xi)'], color='red', label='Эмпирическое распределение f(xi)', s=40)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Тест 3.2.1', fontsize=16)
plt.legend(fontsize=16)
plt.grid(True, alpha=0.5)
plt.show()
```
