```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("test311.csv", sep=",")
print(df.info())
plt.figure(figsize=(16, 9))
#plt.xlim(-10, 10)
plt.plot(df['x'], df['f(x)'], linewidth=2, label='  \nf(x) nu=1, lambda=1, mu=0', alpha=0.8)
plt.scatter(df['xi'], df['f(xi)'], color='red', label='  f(xi)', s=40)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(' 3.1.1', fontsize=16)
#plt.text(0.02, 0.98,'lambda=1', transform=plt.gca().transAxes, fontsize=12,
#         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
plt.legend(fontsize=16)
plt.grid(True, alpha=0.5)
plt.show()
```
