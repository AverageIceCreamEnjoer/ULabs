## Графики распределения

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../dist.csv", sep=',')
```

```python
dist1 = data['dist1']
dist2 = data['dist2']
mixture_dist = data['mixt']
empiric_dist = data['emp']
```

```python
r = list(range(50))
s = 0
for i in range(50): 
  r[i] = s
  s+=0.5
```

```python
data.plot(alpha=0.5)
```

```python
plt.figure(figsize=(15,8))
plt.hist(dist1, bins=r, label='dist1', color='blue')
plt.hist(dist2,bins=r, alpha=0.7, label='dist2', color='red')
plt.hist(mixture_dist, bins=r, alpha=0.7, label='mixture', color='orange')
plt.hist(empiric_dist, bins=r, alpha=0.7, label='empiric', color='green')
plt.xlim(0, 15)
plt.legend(fontsize=16)
```
