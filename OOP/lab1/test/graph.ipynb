## Графики симметричного гиперболического распределения с параметрами $\nu, \mu\space и \space \lambda$

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../dist.csv", sep=',')
```

```python
dist1 = data['dist1']
dist2 = data['dist2']
mixture_dist = data['mixt']
empiric_dist = data['emp']
```

```python
r = list(range(200))
s = -50
for i in range(200): 
  r[i] = s
  s+=0.5
```

### График распределений при $\mu=0, \lambda=1$
Размер выборки для каждого распределления = 10000
- dist1 - основное распределение при $\nu=0.1$
- dist2 - основное распределение при $\nu=30$
- mixt - смесь распределений dist1 и dist2 при $p = 0.5$
- emp - эмпирическое распределение построенное из 10000 случайных величин dist1

```python
data.plot(xlim=(-10, 10), kind='kde', alpha=0.5)
```

```python
plt.figure(figsize=(15,8))
plt.xlim(-15, 15)
plt.hist(dist1, bins=r, label='dist1', color='blue', density=True)
plt.hist(dist2,bins=r, alpha=0.4, label='dist2', color='red', density=True)
plt.hist(mixture_dist, bins=r, alpha=0.4, label='mixture', color='orange', density=True)
plt.hist(empiric_dist, bins=r, alpha=0.6, label='empiric', color='green', density=True)
plt.legend(fontsize=16)
```
